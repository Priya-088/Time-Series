```python
!pip install yfinance
import yfinance
df_tesla = yfinance.download (tickers = "TSLA", start = "2017-02-27",
                              end = "2022-03-13", interval = "1D", group_by = 'ticker', auto_adjust = True, threads = True)
```

```python
#limiting x and y

df_tesla['High'].plot(xlim=['2020-01-01','2021-09-01'], ylim=[0,350],figsize=(12,4))
```

```python
#how to get the indices between '2020-01-01':'2021-09-01'

index = df_tesla.loc['2020-01-01':'2021-09-01'].index
index
```

```python
share_open = df_tesla.loc['2020-01-01':'2021-09-01']['Open']
```

```python
share_open
```

```python
#subplot representation

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
figure,axes=plt.subplots()
axes.plot(index,share_open)
```

```python
#to display x axis properly

figure, axis = plt.subplots()
plt.tight_layout() #to adjust the x axis padding
axis.plot(index,share_open)
```

```python
figure, axis = plt.subplots()
plt.tight_layout()
figure.autofmt_xdate() #preventing the overlapping of the dates

axis.plot(index,share_open)
```
